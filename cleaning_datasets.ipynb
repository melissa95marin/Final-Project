```python
import pandas as pd
import numpy as np
```

### Country Codes

### Hapiness Dataset

```python
hapiness_df= pd.read_csv('Resources/hapiness_data.csv')
hapiness_df
```

```python
#Check how many missing values there are in the dataset
hapiness_df.rename(columns=({'Country name': 'Country_name'}), inplace= True)
hapiness_df.isnull().sum().sum()
```

```python
# Replace null values with 0
hapiness_df= hapiness_df.replace(np.nan, 0)
hapiness_df.isnull().sum().sum()
```

```python
# Adding Country_code to Happiness dataframe
happiness_final_df=hapiness_df.merge(country_codes, how='left', on='Country_name')
happiness_final_df=happiness_final_df[['Country_code', 'Country_name','year','Life Ladder', 'Log GDP per capita','Social support', 'Healthy life expectancy at birth','Freedom to make life choices', 'Generosity','Perceptions of corruption','Positive affect', 'Negative affect', 'Confidence in national government']]
happiness_final_df
```

```python
happiness_final_df.dropna()
```

```python
happiness_final_df.to_csv('Clean_data/Hapiness_final.csv', index=None)
```

### Mortality Dataset ###

```python
mortality= pd.read_csv('Resources/Mortality.txt', header=0 ,sep=',', names=['Country_code', 'Country_name', 'Year', 'Time', 'time_unit','Deaths'])
mortality= mortality.astype({'Country_code':'str', 'Country_name':'str', 'Year':'int', 'Time':'int', 'time_unit':'str','Deaths':'int'})
mortality
```

```python
mortality.loc[(mortality['Country_name'] == 'China')]
```

```python
mortality['Deaths']=mortality['Deaths'].fillna(0)
mortality
```

```python
df=mortality.groupby(['Country_code','Country_name', 'Year'])['Deaths'].sum().reset_index()
print(df)
```

```python
df.isnull().sum().sum()
```

```python
mortality_final=pd.merge(df, country_codes, how='outer',on='Country_code')

mortality_final
```

```python
mortality_final.isnull().sum().sum()
```

```python
mortality_final=mortality_final.dropna()
mortality_final=mortality_final.drop('Country_name_y', axis=1)
mortality_final.rename(columns=({'Country_name_x' : 'Country_name' }),inplace= True)
mortality_final=mortality_final.astype({'Year':'int','Deaths':'int'})
mortality_final
```

```python
mortality_final.to_csv('Clean_data/mortality_final.csv', index=None)
```

### COVID 19 Deaths Dataset ###

```python
covid19=pd.read_csv('Resources/COVID-19global-data.csv', header=0, names= ['year','Alpha_2code','Country_name','Total_cases','number_deaths'],usecols=[1,2,0,4,6], index_col=False )
covid19.head()
covid19.info()
```

```python
covid19['number_deaths'] = covid19['number_deaths'].astype(int)
covid19['year']= pd.to_datetime(covid19['year'])
covid19['year'] =covid19.year.dt.year
```

```python
df2=covid19.groupby(['Country_name','Alpha_2code','year']).agg({'Total_cases':'sum', 'number_deaths':'sum'}).reset_index()
df2=df2[df2.Country_name!='Other']
df2
```

```python
df2=df2.merge(country_codes, how='left', on='Alpha_2code')
df2
```

```python
df2= df2.rename(columns={'Country_name_x':'Country_name'})
df2=df2.drop(['Alpha_2code','Country_name_y' ], axis=1)
```

```python
df2.isnull().sum().sum()
df2
```

```python
covid19_final=df2.dropna()
covid19_final=covid19_final[['Country_code','Country_name','year','Total_cases', 'number_deaths']]
covid19_final.head()
```

```python
covid19_final.to_csv('Clean_data/covid19_final.csv', index=None)
```

### World Population Dataset

```python
df3=pd.read_csv('Resources/WorldPopulation.csv',skiprows=4 ,sep=',', usecols=[0,1, *range(49,66)])
population_df = pd.melt(df3, id_vars=['Country Name', 'Country Code'], var_name='year', value_name='population')
population_df.head()
```

```python
population_df.isnull().sum().sum()
population_df= population_df.dropna()
```

```python
population_df.rename(columns=({'Country Code': 'Country_code'}), inplace= True)
```

```python
population_df=pd.merge( population_df,country_codes, how='outer',on='Country_code')
population_df
```

```python
#population_df.to_csv('test.csv', index=None)
population_df.isnull().sum().sum()
```

```python
population_df= population_df.dropna()
population_df= population_df[['Country_code','Country Name','year', 'population']]
population_df['population'] = population_df['population'].astype(int)
population_df
```

```python
population_df.to_csv('Clean_data/population_final.csv', index=None)
```
